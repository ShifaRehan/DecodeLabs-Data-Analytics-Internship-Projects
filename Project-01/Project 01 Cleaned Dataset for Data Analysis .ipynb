# Check Missing Values

```python
df.isnull().sum()
```

```python
df["CouponCode"] = df["CouponCode"].fillna("NOT APPLIED")
```

```python
df.isnull().sum()
```

# Check for Duplicates

```python
df.duplicated().sum() # No duplicate rows
```

```python
df["OrderID"].duplicated().sum() # No duplicate OrderIDs
```

# Standardize Date Format

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.info()
```

# Remove Extra Spaces

```python
df["CouponCode"] = df["CouponCode"].str.strip()
```

```python
df.head(100)
```

# Clean Text Columns

```python
text_columns = ["Product","ShippingAddress","PaymentMethod","OrderStatus","CouponCode","ReferralSource"]

for col in text_columns:
    df[col] = df[col].str.strip()
```

```python
df.head(10)
```

# OrderID Check

```python
duplicate_orders = df["OrderID"].duplicated().sum()
print("Duplicate Order IDs:", duplicate_orders)
```
